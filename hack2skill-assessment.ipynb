# Q.2

```python
domain_leaders =  pd.read_excel('/kaggle/input/hack2skill-assessment-data/Data Management Task.xlsx', sheet_name="Domain Leaders")
```

```python
domain_leaders.head()
```

```python
domain_leaders_who_responded = pd.merge(
    domain_leaders, 
    feedbacks, 
    how='inner', 
    left_on='Email', 
    right_on='Email Address'
)
domain_leaders_who_responded
```

```python
domain_helpers_who_responded = pd.merge(
    domain_helpers, 
    feedbacks, 
    how='inner', 
    left_on='Email', 
    right_on='Email Address'
)
domain_helpers_who_responded
```

```python
all_respondees = pd.concat([domain_leaders_who_responded[['Name', 'Email']], domain_helpers_who_responded[['Name', 'Email']]]).drop_duplicates()
all_respondees
```

```python
discord_records = pd.read_excel('/kaggle/input/hack2skill-assessment-data/Data Management Task.xlsx', sheet_name="Discord Entries")
discord_records
```

```python
respondees_with_discord = pd.merge(
    all_respondees, 
    discord_records, 
    how='inner', 
    left_on='Email', 
    right_on='Email Address'
)
respondees_with_discord[["Name", "Email","Discord Username" ]]
```

# Q.3

```python
discord_score = pd.read_excel('/kaggle/input/hack2skill-assessment-data/Data Management Task.xlsx', sheet_name="Best Discord Performer")
discord_score
```

```python
all_members = pd.concat([domain_leaders, domain_helpers]).drop_duplicates().reset_index()
all_members
```

```python
all_members_with_discord_username = pd.merge(
    all_members, 
    discord_records, 
    how="inner", 
    left_on='Email', 
    right_on='Email Address'
)[['Name', 'Email', 'Discord Username']]

all_members_with_discord_username
```

```python
discord_member_details = pd.merge(
    discord_score[['Discord IDs']], 
    all_members_with_discord_username[['Discord Username', 'Email', 'Name']], 
    how="inner", 
    left_on='Discord IDs', 
    right_on='Discord Username'
)[['Name', 'Email']]

discord_member_details
```
